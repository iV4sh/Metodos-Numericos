### **Graficar Funciones**
$y(x) = x^3 - 7x^2 + 4x + 12$

* **Tabla de valores**

| i   | x   | y   |
|-------------|-------------|-------------|
| 0    | -5    | -308    |
| 1    | 0    | 12    |
| 2    | 5    | -18    |


```python
# Definici√≥n de la ecuaci√≥n
def eqn(x):
    return x**3 - 7 * x**2 + 4 * x + 12
```

```python
# Generador de puntos para la gr√°fica
import numpy as np

## Generar 100 datos del -10 al 10
Xs, step = np.linspace(-10, 10,num=100, retstep=True)
```

```python
# Generador de valores
Ys = []

for x in Xs:
    y = eqn(x)
    Ys.append(y)
    print(f"x = {x:.2f} \t y = {y:.2f}")
```

```python
from matplotlib import pyplot as plt
# Alternative: import seaborn as sns
plt.plot(Xs, Ys, marker=".",linestyle=":")
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title("Grafica de la funci√≥n c√∫bica")
plt.grid()
plt.legend(["$y(x) = x^3 - 7x^2 + 4x + 12$"])


## Tarea hacer zoom

plt.xlim(-5, 8)  # Zoom en el eje x
plt.ylim(-25, 20) # Zoom en el eje y

plt.show()
```
